```python
import pandas as pd
import seaborn as sn
```

```python
dados = pd.read_csv('licitacoesAcaraú.csv')
dados.head()
```

```python
dados.info()
```

```python
dados.isna().sum()
```

```python
dados['Valor estimado'] = dados['Valor estimado'].str.replace(".", "", regex=True)
dados['Valor estimado'] = dados['Valor estimado'].str.replace(",", ".", regex=True)
dados['Valor estimado'] = dados['Valor estimado'].astype(float)
```

```python
dados.info()
```

```python
dados.loc[dados['Valor estimado'].isna(), 'Valor estimado'] = dados['Valor estimado'].median()
```

```python
dados.info()
```

```python
df_num = dados.select_dtypes(include=('object')).columns
for c in df_num:
    dados[c].fillna(dados[c].mode().values[0], inplace=True)

dados.info()
```

```python
secretarias = {
    "CONTROLADORIA GERAL DO MUNICÍPIO E OUVIDORIA": ["CONTROLADORIA", "OUVIDORIA", "TRANSPARÊNCIA", "FISCALIZAÇÃO"],
    "GABINETE DO PREFEITO": ["GABINETE", "PREFEITO", "GESTÃO MUNICIPAL", "PLANEJAMENTO ESTRATÉGICO"],
    "PROCURADORIA GERAL DO MUNICÍPIO": ["PROCURADORIA", "JURÍDICO", "ADVOCACIA", "ASSESSORIA JURÍDICA", "PROCESSO JUDICIAL"],
    "SECRETARIA DE ADMINISTRAÇÃO E FINANÇAS": ["RECURSOS HUMANOS","UNIDADES ADMINISTRATIVAS","DIVERSAS UNIDADES ADMINISTRATIVAS","ADMINISTRAÇÃO", "FINANÇAS", "ORÇAMENTO", "CONTABILIDADE", "TRIBUTOS", "TRIBUTÁRIA", "LICITAÇÕES", "LICITAÇÃO", "FOLHA SALARIAL", "SALÁRIO"],
    "SECRETARIA DE AGRONEGÓCIOS, PESCA, IRRIGAÇÃO E DESENVOLVIMENTO ECONÔMICO": ["AGRONEGÓCIOS", "PESCA", "IRRIGAÇÃO", "AGRICULTURA", "PECUÁRIA", "DESENVOLVIMENTO ECONÔMICO"],
    "SECRETARIA DE ASSISTÊNCIA SOCIAL": ["ASSISTENCIA SOCIAL", "ASSISTÊNCIA SOCIAL", "CRAS", "CREAS", "FAMÍLIAS CARENTES", "INCLUSÃO SOCIAL", "BOLSA FAMÍLIA"],
    "SECRETARIA DE EDUCAÇÃO": ["EDUCAÇÃO", "ESCOLA", "ALUNOS", "PROFESSORES", "MERENDA ESCOLAR", "CRECHE", "ENSINO"],
    "SECRETARIA DE ESPORTE E JUVENTUDE": ["ESPORTE", "JUVENTUDE", "LAZER", "ATIVIDADE FÍSICA", "CAMPEONATO"],
    "SECRETARIA DE INFRAESTRUTURA": ["DRENAGEM", "VEÍCULOS", "FROTA", "ESTRADA", "ESTRADAS","RECAPEAMENTO", "ASFALTO","RESTAURAÇÃO", "RESTAURAÇÕES", "INFRAESTRUTURA", "INTRAESTRUTURA","INFRA-ESTRUTURA", "OBRAS", "PAVIMENTAÇÃO", "ENGENHARIA", "SANEAMENTO", "URBANIZAÇÃO", "REFORMA", "REFORMAS", "MANUTENÇÃO", "MANUTENÇÕES", "CONSTRUÇÃO", "CONSTRUÇÕES"],
    "SECRETARIA DE MEIO AMBIENTE": ["MEIO AMBIENTE", "SUSTENTABILIDADE", "RESÍDUOS", "LICENCIAMENTO AMBIENTAL", "PRESERVAÇÃO"],
    "SECRETARIA DE SAÚDE": ["SAÚDE", "SAMU", "PACIENTE", "HOSPITAL", "POSTO DE SAÚDE", "AMBULÂNCIA", "MEDICAMENTOS", "VACINA", "ATENDIMENTO MÉDICO", "CAPS", "CLÍNICA"],
    "SECRETARIA DE SEGURANÇA E TRÂNSITO": ["SEGURANÇA", "TRÂNSITO", "GUARDA MUNICIPAL", "VIATURA", "FISCALIZAÇÃO DE TRÂNSITO"],
    "SECRETARIA DE TURISMO E CULTURA": ["TURISMO", "CULTURA", "EVENTOS CULTURAIS", "PATRIMÔNIO HISTÓRICO", "FESTIVAIS"],
    "SECRETARIA MUNICIPAL DE RELAÇÕES INSTITUCIONAIS E POLÍTICAS": ["RELAÇÕES INSTITUCIONAIS", "POLÍTICAS PÚBLICAS", "ARTICULAÇÃO POLÍTICA"],
    "SECRETARIAS DIVERSAS":["DIVERSAS SECRETARIAS", "AS SECRETARIAS"]
}
# Função para classificar a secretaria
def identificar_secretaria(texto):
    for secretaria, palavras in secretarias.items():
        if any(palavra in texto.upper() for palavra in palavras):
            return secretaria
    return "Não Identificada"  # Caso não encontre correspondência

# Criando a coluna 'Secretaria'
dados["Secretaria"] = dados["Objeto"].apply(identificar_secretaria)
```
